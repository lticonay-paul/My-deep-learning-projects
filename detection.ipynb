```python
# Modelo de deteccion
# yolo 8, pytorch
# Las imagenes de entrada en la carpeta "images"
# Las etiquetas de salida en la carpeta "labels"
# En images estan las carpetas train, val
# En labels estan las carpetas train, val

import glob

# Limpiar etiquetas: todas las clases se cambian a 0 y solo se mantienen líneas válidas (formato YOLO)
for split in ['train', 'val']:
    label_files = glob.glob(f'/media/ferdinand/Datos/dataset/student/labels/{split}/*.txt')
    for file in label_files:
        with open(file, 'r') as f:
            lines = f.readlines()
        new_lines = []
        for line in lines:
            parts = line.strip().split()
            # Solo procesa líneas con al menos 5 valores (clase + bbox)
            if len(parts) >= 5:
                parts[0] = '0'
                new_lines.append(' '.join(parts) + '\n')
        with open(file, 'w') as f:
            f.writelines(new_lines)

with open('data.yaml', 'w') as f:
    f.write(
        "train: /media/ferdinand/Datos/dataset/student/images/train\n"
        "val: /media/ferdinand/Datos/dataset/student/images/val\n"
        "nc: 1\n"
        "names: ['clase0']\n"
    )

from ultralytics import YOLO

# Cargar el modelo YOLOv8 preentrenado
model = YOLO("yolov8n.pt")

data_yaml = 'data.yaml'  # Asegúrate de tener este archivo en tu carpeta de trabajo
# Selección automática de dispositivo para entrenamiento
device_train = 0 if torch.cuda.is_available() else 'cpu'
print(f"Entrenando en: {device_train}")

results = model.train(
    data=data_yaml,
    epochs=2,
    imgsz=640,
    batch=8,
    workers=2,
    device=device_train,
    amp=False,
    optimizer='SGD'  # Cambia a SGD
)

# Guardar el modelo entrenado
model.save('yolov8n_custom.pt')

import torch
from ultralytics import YOLO

# Seleccionar dispositivo automáticamente
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Usando dispositivo: {device}")

# Cargar el modelo entrenado
model = YOLO('yolov8n_custom.pt')

test_images = '/media/ferdinand/Datos/dataset/student/images/val'  # O una imagen individual: 'images/val/ejemplo.jpg'

results = model.predict(source=test_images, save=True, imgsz=640, conf=0.25, device=device)

for r in results:
    r.show()  # Muestra la imagen con las detecciones
# Las imágenes con las detecciones se guardan en 'runs/detect/predict'
```
